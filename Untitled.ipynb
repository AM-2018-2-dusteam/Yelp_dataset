```python
def balance(reviews,size):
    stars_5 = (reviews[reviews.stars == 5])
    stars_4 = (reviews[reviews.stars == 4])
    stars_3 = (reviews[reviews.stars == 3])
    stars_2 = (reviews[reviews.stars == 2])
    stars_1 = (reviews[reviews.stars == 1])
    
    assert (len(stars_1) >= size) 
    assert (len(stars_2) >= size)
    assert (len(stars_3) >= size)
    assert (len(stars_4) >= size)
    assert (len(stars_5) >= size)
    
    stars_5 = stars_5.iloc[:size]
    stars_4 = stars_4.iloc[:size]
    stars_3 = stars_3.iloc[:size]
    stars_2 = stars_2.iloc[:size]
    stars_1 = stars_1.iloc[:size]
      
    return pd.concat([stars_1,stars_2,stars_4,stars_5])
```

```python
reviews = balance(reviews,1000)
reviews = reviews.reset_index()
```

```python
reviews = reviews.drop(columns=['business_id', 'date','review_id','user_id','stars'])
```

```python

def text_process(text):

    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
#reviews['text'] = reviews['text'].apply(text_process)
#vec = CountVectorizer(analyzer=text_process).fit(reviews['text'])
vec = CountVectorizer(analyzer=text_process,tokenizer=None,preprocessor=None,stop_words=None,max_features=5000)
```

```python
vec.fit(reviews['text'])
X = vec.transform(reviews['text'])
```

```python
X = X.toarray()
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, stars, test_size=0.33, random_state=42)
```

```python
nb = MultinomialNB()
#nb = RandomForestClassifier(n_estimators=100)
nb.fit(X_train, y_train)

```

```python
nb.score(X_train, y_train)
```

```python
nb.score(X_test, y_test)
```

```python

#Confusion Matrixfor Test Data
conf_array_test = confusion_matrix(y_test,nb.predict(X_test))
conf_array_test =  conf_array_test / conf_array_test.astype(np.float).sum(axis=1)
df_class_test = pd.DataFrame(conf_array_test, range(2),range(2))
sn.set(font_scale=0.7)#for label size
sn.heatmap(df_class_test,annot=True)# font size
```
